```python
import plotly.express as px
import cufflinks as cf
cf.go_offline()
import plotly.graph_objects as go
```

```python
#COUNTRY WISE ANALYSIS OF THE PLAYERS
```

```python
rankings = pd.read_csv('fifa rankings updated.csv')
```

```python
#TOP 50 RANKED COUNTRIES BY FIFA
rankings.drop(rankings.columns[[0]], axis = 1, inplace = True)
rankings = rankings.reset_index(drop=True)
rankings
```

```python
rankings.to_csv('fifa rankings updated.csv')
```

```python
#_AGE V/S POTENTIAL plotting for different players of the top 10 ranked countries
i=0
while i < 10:
    df = dataset.loc[np.where(dataset['Nationality']== rankings['Team'][i])[0]]
    fig = px.scatter(df, x='Age', y='Potential', hover_name="Name",title = rankings['Team'][i], color ='Best Position' )
    fig.show()
    i+=1                 
```

```python
#BLANK DATAFRAME
df = pd.DataFrame(columns = dataset.columns)
```

```python
df
```

```python
#ANALYSING YOUNG PLAYERS AND THEIR POTENTIAL
i=0
while i < 30:
    df = pd.concat([dataset[(dataset["Age"] <= 23) & (dataset["Potential"] > 80) & (dataset['Nationality']== rankings['Team'][i])],df], ignore_index=True)
    i+=1
```

```python
df
```

```python
df.to_csv('youngsters top30 nations.csv')
```

```python
fig = px.bar(df, x='Nationality', y='Potential', hover_name="Name",title = 'YOUNGSTERS', color ='Best Position',barmode = 'group')
fig.show()
```

```python
fig = px.sunburst(df, path=['Nationality', 'Name'], values='Potential',
                  color='Best Position', hover_data=['Age'])
fig.show()
```

```python
df['Best Position'].unique()
```

```python
CAM = df[(df["Best Position"]=="CAM")].sort_values(by="Potential",ascending = False).head(10)
```

```python
CAM
```

```python
midfield_attributes =[] 
midfield_attributes= ['Crossing','ShortPassing', 'Reactions','Dribbling','LongPassing','BallControl','Balance','Stamina','Strength','Vision','Positioning','Interceptions']
```

```python
CAM.loc[(CAM['Best Position']=='CAM'),midfield_attributes]
```

```python
for i, row in CAM.loc[(CAM['Best Position']=='CAM'),midfield_attributes].head(5).iterrows():
    #print (df['Name'][i])
    fig = go.Figure()
    for i, row in CAM.loc[(CAM['Best Position']=='CAM'),midfield_attributes].head(5).iterrows():
        fig.add_trace(go.Scatterpolar(r=row , theta= midfield_attributes,fill = 'toself',name = df['Name'][i] ,showlegend=True))
fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0, 100])),showlegend=True)
fig.show()
```

```python
for i, row in CAM.loc[(CAM['Best Position']=='CAM'),midfield_attributes].tail(5).iterrows():
    #print (df['Name'][i])
    fig = go.Figure()
    for i, row in CAM.loc[(CAM['Best Position']=='CAM'),midfield_attributes].tail(5).iterrows():
        fig.add_trace(go.Scatterpolar(r=row , theta= midfield_attributes,fill = 'toself',name = df['Name'][i] ,showlegend=True))
fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0, 100])),showlegend=True)
fig.show()
```
